### 1. A recipe calls for 3 cups of flour for every 4 cups of sugar. How much sugar is needed if 6 cups of flour are used?

3 cups of flour / 4 cups of sugar = 6 cups of flour / x cups of sugar

To find the value of "x," we can cross-multiply:

3 cups of flour * x cups of sugar = 6 cups of flour * 4 cups of sugar

3x = 24

Dividing both sides by 3:

x = 8

### 2. An art piece is sold for $120, which is a 20% profit over the cost. What was the cost price?

Selling Price = Cost Price + Profit

$120 = C + 20% of C

To calculate 20% of C, we multiply C by 0.20:

$120 = C + 0.20C

Combining like terms:

$120 = 1.20C

Now, we can solve for C by dividing both sides of the equation by 1.20:

C = $120 / 1.20

C = $100

### 3. A rectangular garden has a length of 20 meters and a width that is half of its length. What is the area of the garden?

the length of the garden is 20 meters, we can determine the width by dividing the length by 2:

Width = Length / 2 = 20 / 2 = 10 meters

Now, we can calculate the area:

Area = Length × Width = 20 × 10 = 200 square meters

### 13. Calculate the surface area of a cylinder with a radius of 3 units and a height of 5 units.

The curved surface area of a cylinder is given by the formula:
Cylinder CSA = 2 * π * radius * height

The area of each base of the cylinder is given by the formula:
Cylinder Base Area = π * radius^2

Radius = 3 units
Height = 5 units

Calculating the curved surface area:
Cylinder CSA = 2 * π * 3 * 5
Cylinder CSA = 30 * π

Calculating the area of each base:
Cylinder Base Area = π * 3^2
Cylinder Base Area = 9 * π

The total surface area of the cylinder is the sum of the curved surface area and the area of the two bases:
Cylinder Surface Area = Cylinder CSA + 2 * Cylinder Base Area
Cylinder Surface Area = 30 * π + 2 * (9 * π)

Simplifying:
Cylinder Surface Area = 30π + 18π
Cylinder Surface Area = 48π

Therefore, the surface area of the cylinder with a radius of 3 units and a height of 5 units is 48π square units.

### 15. If you have a square plot of land with an area of 49 square meters, what is the perimeter of the plot?

The area of the square plot is 49 square meters, we can find the length of one side by taking the square root of the area.

√49 = 7 meters

Since a square has four equal sides, we can multiply the length of one side by 4 to find the perimeter:

Perimeter = 4 * 7 meters

Perimeter = 28 meters

The square plot of land with an area of 49 square meters is 28 meters
