### 1. Convert 50 centimeters to inches.

   1 inch is equal to 2.54 centimeters.

So, to convert 50 centimeters to inches, we divide 50 by 2.54:

50 cm ÷ 2.54 cm/inch = 19.685 inches

### 2. What is the perimeter of a rectangle with length 6 units and width 4 units?

To find the perimeter, we can use the formula:

Perimeter = 2 * (length + width)

Perimeter = 2 * (6 + 4)
Perimeter = 2 * 10
Perimeter = 20

### 3. Calculate the area of a triangle with base 5 units and height 3 units.

The area of a triangle can be calculated using the formula:

Area = (base * height) / 2

Area = (5 * 3) / 2
Area = 15 / 2
Area = 7.5

### 5. What is the circumference of a circle with a radius of 4 units?

Circumference = 2 * π * 4

Using an approximate value of π ≈ 3.14159, we can calculate:

Circumference ≈ 2 * 3.14159 * 4
Circumference ≈ 6.28318 * 4
Circumference ≈ 25.13272

Therefore, the circumference of the circle with a radius of 4 units is approximately 25.13272 units.
