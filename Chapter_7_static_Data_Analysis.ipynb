### 8. If a coin is flipped three times, what is the probability of getting exactly two heads?

When flipping a fair coin, the probability of getting a head or a tail on any given flip is 1/2 or 0.5.

To calculate the probability of getting exactly two heads when flipping a coin three times, we need to consider the number of ways this outcome can occur.

In this case, there are three possible outcomes where exactly two heads occur: HHT, HTH, and THH. The remaining outcome would be TTT.

Since each flip is independent, the probability of a specific outcome occurring is the product of the probabilities of each individual flip. Therefore, the probability of each of the three outcomes with exactly two heads is:

Probability of HHT = (1/2) * (1/2) * (1/2) = 1/8
Probability of HTH = (1/2) * (1/2) * (1/2) = 1/8
Probability of THH = (1/2) * (1/2) * (1/2) = 1/8

Adding up the probabilities of these three outcomes:

Probability of exactly two heads = (1/8) + (1/8) + (1/8) = 3/8

Therefore, the probability of getting exactly two heads when flipping a coin three times is 3/8 or approximately 0.375.
