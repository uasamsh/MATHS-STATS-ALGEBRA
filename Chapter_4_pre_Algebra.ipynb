### 1. Convert the fraction 3/5 to a percentage.
   3/5 * 100 = (3 * 100) / 5 = 300/5 = 60
   the fraction 3/5 is equivalent to 60%.

### 2. Change 40% to a decimal.

### 4. If a dress costs $70 after a 30% discount, what was the original price?

The original price as "x".
After a 30% discount, the discounted price is 70 dollars, which is equal to 70% of the original price. 
70% of x = $70
0.70x = $70
x = $70 / 0.70
x = $100
The original price of the dress was $100.  

### 5. A population of 50,000 increases by 5% per year. What is the population after one year?
   5% = 5/100 = 0.05
   The increase in population over one year is 0.05*50000 = 2500 people.
   So, the population after one year would be 50000 + 2500 = 52500.
